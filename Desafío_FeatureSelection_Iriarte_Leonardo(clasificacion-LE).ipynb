Modificamos los valores totales de la columna 'Onboard_Total' ya que tiene un formato repetitivo respecto a las 2 columnas anteriores y la posterior.
Nos será de mayor utilidad si sólo mostrara un valor numérico de total de pasajeros y tripulación para luego hacer cálculos precisos con respecto a la columna 'Fatalities', que también es tipo int64.

```python
def procesar_valor(valor):

    partes = valor.split('/ ')
    if len(partes) > 1:
        valor_final = partes[1]
    else:
        valor_final = partes[0]

    numeros = [int(s) for s in valor_final.split() if s.isdigit()]

    if len(numeros) > 0:
        return numeros[-1]
    else:
        return 0

df['Onboard_Total'] = df['Onboard_Total'].apply(procesar_valor)
```

Se aplica un código que convierta éstos datos, al igual que los anteriores, para utilizar las columnas en formato int. Para ello, se deja sólo el número de personas a bordo en cada columna extrayendo el número final de cada registro y en caso de no tener número, se asigna 0.

Además se crean 2 columnas nuevas llamadas 'Crew_Fatalities' y 'Pass_Fatalities' las cuales utilizan otro proceso de extracción de número, pero en éste caso el que se asigna luego de 'Fatalities: ' y y en caso de no tener número, se asigna 0.

```python
def extract_fatalities(s):
    numbers = re.findall(r'\d+', s)
    return int(numbers[0]) if numbers else 0

df['Crew_Fatalities'] = df['Onboard_Crew'].apply(lambda x: extract_fatalities(x.split(' / ')[0]))
df['Pass_Fatalities'] = df['Onboard_Passengers'].apply(lambda x: extract_fatalities(x.split(' / ')[0]))
df['Onboard_Crew'] = df['Onboard_Crew'].apply(lambda x: int(re.findall(r'\d+', x.split(' / ')[1])[0]) if re.findall(r'\d+', x.split(' / ')[1]) else 0)
df['Onboard_Passengers'] = df['Onboard_Passengers'].apply(lambda x: int(re.findall(r'\d+', x.split(' / ')[1])[0]) if re.findall(r'\d+', x.split(' / ')[1]) else 0)

df['Onboard_Crew'] = df['Onboard_Crew'].astype(int)
df['Onboard_Passengers'] = df['Onboard_Passengers'].astype(int)
df['Crew_Fatalities'] = df['Crew_Fatalities'].astype(int)
df['Pass_Fatalities'] = df['Pass_Fatalities'].astype(int)

df[['Onboard_Crew', 'Crew_Fatalities', 'Onboard_Passengers', 'Pass_Fatalities']].head(10)
```

Una vez creada la columna 'Year' podemos utilizarla para crear la columna previamente mencionada 'Aircraft_Age'.

```python
def calcular_edad(row):
    first_flight_year = str(row['Aircaft_First_Flight'])[:4]
    if first_flight_year.isdigit() and int(first_flight_year) != 0:
        return row['Year'] - int(first_flight_year)
    else:
        return 0

df['Aircraft_Age'] = df.apply(calcular_edad, axis=1).astype(int)

df['Aircraft_Age'].value_counts().head(10)
```

Agregamos la columna 'Flight_Type' la cual define con una función condicional si el país de la columna 'Departure_Airport' coincide con el país de la columna 'Destination_Airport' entonces devolverá el valor 'Domestic', caso contrario el valor será 'International' y se asignará 'Unknwon' a los registros que poseen datos como '?' y '-'.

```python
print(df['Departure_Airport'].value_counts(ascending=False).head(5))
print('-------------------------------------------------------------')
print(df['Destination_Airport'].value_counts(ascending=False).head(5))
```

```python
def determine_flight_type(row):
    if ',' in row['Departure_Airport'] and ',' in row['Destination_Airport']:

        departure_after_comma = row['Departure_Airport'].split(',')[1].strip()
        destination_after_comma = row['Destination_Airport'].split(',')[1].strip()

        if '?' in departure_after_comma or '?' in destination_after_comma or '-' in departure_after_comma or '-' in destination_after_comma:
            return 'Unknown'
        elif departure_after_comma == destination_after_comma:
            return 'Domestic'
        else:
            return 'International'
    else:
        return 'Unknown'

df['Flight_Type'] = df.apply(determine_flight_type, axis=1)

df['Flight_Type'].head(10)
```

La columna 'Incident_Category' tiene 10 variables categóricas, las cuales vamos a agrupar en 2 que generalicen y separen accidentes de hechos criminales, creando una nueva columna llamada 'Category':

VARIABLE 'ACCIDENTS'.

- **Accident | repairable-damage:** Accidente; daño reparable.
- **Accident | hull-loss:** Accidente; pérdida del casco.
- **Incident | repairable-damage:** Incidente; daño reparable.
- **Occurrence unknown | hull-loss:** Suceso desconocido; pérdida del casco.

VARIABLE 'CRIMINAL'.

- **Criminal occurrence (sabotage, shoot down) | hull-loss:** Hecho criminal (sabotaje, derribo); pérdida del casco.
- **Other occurrence (ground fire, sabotage) | hull-loss:** Otros sucesos (incendio terrestre, sabotaje); pérdida del casco.
- **Criminal occurrence (sabotage, shoot down) | repairable-damage:** Hecho criminal (sabotaje, derribo); daño reparable.
- **Hijacking | repairable-damage:** Secuestro | daño reparable.
- **Other occurrence (ground fire, sabotage) | repairable-damage:** Otros sucesos (incendio terrestre, sabotaje); daño reparable.
- **Hijacking | hull-loss:** Secuestro; pérdida del casco

```python
df['Incident_Category'].value_counts()
```

```python
def determine_category(row):
    incident_category = row['Incident_Category']

    if any(category in incident_category for category in ['Accident | repairable-damage', 'Accident | hull-loss', 'Incident | repairable-damage', 'occurrence unknown | hull-loss','other occurrence (ground fire, sabotage) | repairable-damage']):
        return 'Accidents'
    elif any(category in incident_category for category in ['Criminal occurrence (sabotage, shoot down) | hull-loss', 'other occurrence (ground fire, sabotage) | hull-loss', 'Criminal occurrence (sabotage, shoot down) | repairable-damage', 'Hijacking | repairable-damage', 'Criminal occurrence (sabotage, shoot down) | repairable-damage', 'Hijacking | hull-loss']):
        return 'Criminal'
    else:
        return 'Other'

df['Category'] = df.apply(determine_category, axis=1)

df['Category'].value_counts()
```

Creamos una columna llamada 'Country' la cual se asignan valores según una función que determina si corresponde el país que figura posterior a la coma de las columnas 'Destination_Airport' o 'Departure_Airport', dependiendo de la fase de vuelo registrada en la columna 'Aircraft_Phase'.

```python
df['Aircraft_Phase'].value_counts()
```

```python
def determine_country(row):
    phase = row['Aircraft_Phase']

    if phase in ['Landing (LDG)', 'En route (ENR)', 'Approach (APR)', 'Maneuvering (MNV)']:
        return row['Destination_Airport'].split(', ')[-1]
    elif phase in ['Standing (STD)', 'Takeoff (TOF)', 'Initial climb (ICL)', 'Pushback / towing (PBT)', 'Taxi (TXI)']:
        return row['Departure_Airport'].split(', ')[-1]
    elif phase == 'Unknown (UNK)':
        return 'Unknown'
    else:
        return None

df['Country'] = df.apply(determine_country, axis=1)

df['Country'].value_counts().head(10)
```

```python
def determine_utility(row):
    nature = row['Aircaft_Nature']

    if nature in ['Domestic Non Scheduled Passenger', 'Domestic Scheduled Passenger', 'Non Scheduled Passenger',
                  'Scheduled Passenger', 'Passenger', 'Private', 'Executive',
                  'International Scheduled Passenger', "Int'l Non Scheduled Passenger", 'Official state flight']:
        return 'Passenger'
    elif nature in ['Cargo', 'Test', 'Training', 'Ambulance', 'Fire fighting',
                    'Aerial Work (Calibration, Photo)', 'Survey/research', 'Agricultural', 'Ferry/positioning']:
        return 'Aerial_Work'
    elif nature in ['Illegal Flight', 'Parachuting', 'Demonstration']:
        return 'Other'
    elif nature == 'Military':
        return 'Military'
    elif nature in ['Unknown', '-']:
        return 'Unknown'
    else:
        return 'Unknown'

df['Aircraft_Utility'] = df.apply(determine_utility, axis=1)

df['Aircraft_Utility'].value_counts()
```

También creamos la columna objetivo, la cual se llama 'Survival_Prob' y tiene un cálculo simple de probabilidad entre las personas total a bordo según la columna 'Onboard_Total' y las víctimas fatales según la columna 'Fatalities'.

```python
df['Survival_Prob'] = (df['Onboard_Total'] - df['Fatalities']) / df['Onboard_Total']*100
df['Survival_Prob'] = df['Survival_Prob'].replace([np.nan, np.inf, -np.inf], 0).round().astype(int)

df['Survival_Prob'].unique()
```

## **[MACHINE LEARNING](#machine-learning)**

Para éste dataset utilizaremos un algoritmo de clasificación.\
Comenzamos quitando los registros de aeronaves de uso militar y cuyo incidente se diera durante la 2da guerra mundial (01/09/1939 - 02/09/1945). Dichos datos no se borran, sino que quedan asignados en un 2do dataframe llamado 'df_ww2'.

```python
condicion_ww2 = (df['Incident_Date'].between('1939-09-01', '1945-09-02')) & (df['Aircaft_Nature'] == 'Military')
df_ww2 = df[condicion_ww2]

df = df[~condicion_ww2]
```

Luego convertimos los valores de la columna 'Survival_Prob' para transformarla en binario, lo cual mejorará el funcionamiento del modelo.

```python
df['Survival_Prob'] = (df['Onboard_Total'] - df['Fatalities']) / df['Onboard_Total']
df['Survival_Prob'] = df['Survival_Prob'].replace([np.nan, np.inf, -np.inf], 0).round().astype(int)

df['Survival_Prob'].value_counts()
```

Creamos un gráfico de barras simple para corroborar la distribución de la columna.

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Survival_Prob'], bins=10, color='b')
plt.xlabel('Probabilidad de Supervivencia')
plt.ylabel('Conteo')
plt.title('Distribución de Probabilidad de Supervivencia')
plt.show()
```

Comenzamos eligiendo las columnas que más influyen en el objetivo a predecir, separándolas en 2 variables para poder convertir las categóricas en numéricas mediante Label Encoder y luego volviendo a concatenarlas con las que ya eran tipo int en un dataset final para entrenar y testear.
