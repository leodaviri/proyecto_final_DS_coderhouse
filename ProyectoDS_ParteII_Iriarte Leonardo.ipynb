A modo de organización, creamos 2 columnas que especifican año y mes de la columna 'Incident_Date':
- Sobre la columna Año, extraemos sólo el año de la columna 'Incident_Date'.
- Sobre la columna Mes, asignamos y mostramos la abreviación textual de cada uno.

```python
df['Year'] = df['Incident_Date'].dt.year
df['Month'] = df['Incident_Date'].dt.month

meses = {
    1: 'Ene',
    2: 'Feb',
    3: 'Mar',
    4: 'Abr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Ago',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dic'}

df['Month'] = df['Month'].map(meses)

df[['Month', 'Year']]
```

Visualización de los años donde se registraron los casos.

```python
df['Year'].unique()
```

Modificamos los valores totales de la columna 'Onboard_Total' ya que tiene un formato repetitivo respecto a las 2 columnas anteriores y la posterior.
Nos será de mayor utilidad si sólo mostrara un valor numérico de total de pasajeros y tripulación para luego hacer cálculos precisos con respecto a la columna 'Fatalities', que también es tipo int64.

```python
def procesar_valor(valor):

    partes = valor.split('/ ')
    if len(partes) > 1:
        valor_final = partes[1]
    else:
        valor_final = partes[0]

    numeros = [int(s) for s in valor_final.split() if s.isdigit()]

    if len(numeros) > 0:
        return numeros[-1]
    else:
        return 0

df['Onboard_Total'] = df['Onboard_Total'].apply(procesar_valor)

df['Onboard_Total']
```

Agregamos la columna 'Flight_Type' la cual define con una función condicional si el país de la columna 'Departure_Airport' coincide con el país de la columna 'Destination_Airport' entonces devolverá el valor 'Domestic', caso contrario el valor será 'International' y se asignará 'nan' a los registros que poseen datos como '?' y '-'.

```python
def extract_country(text):
    match = re.search(r',\s([^,]+)$', text)
    if match:
        return match.group(1)
    else:
        return np.nan
def determine_flight_type(row):
    departure_country = extract_country(row['Departure_Airport'])
    destination_country = extract_country(row['Destination_Airport'])
    if pd.isna(departure_country) or pd.isna(destination_country):
        return np.nan
    if departure_country == destination_country:
        return 'Domestic'
    else:
        return 'International'
df['Flight_Type'] = df.apply(determine_flight_type, axis=1)

df['Flight_Type']
```

Cobre la columna 'Aircaft_First_Flight' se identifican diversidad de formatos de valores, donde a veces posee solo 4 dígitos que indican un año, en otros casos poseen una fecha completa y también pueden tener una referencia de edad por meses y/o años de antigüedad entre parentesis.

```python
df['Aircaft_First_Flight']
```

Ingresamos un código de cálculo que nos permita extraer sólo los primeros 4 dígitos que indiquen el año del primer vuelo y luego restar dicho valor - el número existente en la columna 'Year' para devolver la edad de la aeronave al momento del suceso, en una columna nueva llamada 'Aircraft_Age'. Se agrega un condifinal que devuelva valor 0 en caso de que la columna 'Aircaft_First_Flight' no posea valores, sea NaN o tenga valor 0, para que en ésos casos entonces no realice la resta, sino que devuelva valor 0 en la columna nueva (de ésta manear evitamos posibles errores en los cuales, por ejemplo, veamos aeronaves con edad 1962).

```python
def calcular_edad(row):
    first_flight_year = str(row['Aircaft_First_Flight'])[:4]
    if first_flight_year.isdigit() and int(first_flight_year) != 0:
        return row['Year'] - int(first_flight_year)
    else:
        return 0

# Aplicar la función a cada fila del DataFrame y crear la nueva columna 'Aircraft_Age'
df['Aircraft_Age'] = df.apply(calcular_edad, axis=1).astype(int)

df['Aircraft_Age']
```

Corremos un código de los valores más altos obtenidos en la nueva columna, con el fin de detectar posibles errores en caso de anomalías.

- se utilizan funciones *unique* y *value_counts*.
- dada la cantidad de valores únicos, se utiliza un top 20 de mayor cantidad de repeticiones.

```python
df['Aircaft_Model'].unique().tolist()
```

```python
df['Aircaft_Model'].nunique()
```

```python
df['Aircaft_Model'].value_counts().head(20)
```

```python
top20_acm = df['Aircaft_Model'].value_counts().head(20)
top20_acm.plot.bar()

plt.xlabel('Modelo de Aeronave')
plt.ylabel('Cantidad de casos')
plt.title('Top 20 de Aeronaves Según Modelo')
```

---

Sobre éste top 20 obtenido intentaremos obtener el modelo de aeronave que menor promedio tenga de fallecidos según personas a bordo.

- se utilizan funciones *.sum()* para las columnas 'Onboard_Total' y 'Fatalities'.
- se consideran sólo los modelos de aeronaves según el top 20 obtenido previamente.

```python
data = {
    'Aircaft_Model': top20_acm.index,
    'Onboard_Total_Sum': [df[df['Aircaft_Model'] == model]['Onboard_Total'].sum() for model in top20_acm.index],
    'Fatalities_Sum': [df[df['Aircaft_Model'] == model]['Fatalities'].sum() for model in top20_acm.index]
}

df_top20 = pd.DataFrame(data)
df_top20
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df_top20['Aircaft_Model'], df_top20['Onboard_Total_Sum'], marker='o', label='Total a Bordo')
plt.plot(df_top20['Aircaft_Model'], df_top20['Fatalities_Sum'], marker='o', label='Fallecidos')

plt.xlabel('Modelo de Aeronave')
plt.ylabel('Cantidad de Casos y Fallecidos')
plt.title('Comparación de Fallecidos y Personas a Bordo según Top 20 de Aeronaves')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

Creamos una lista para visualizar la aeronave con mayor cantidad de fallecidos:

```python
df_top20[['Aircaft_Model', 'Onboard_Total_Sum', 'Fatalities_Sum']].sort_values(by='Fatalities_Sum', ascending=False)
```

Veamos éstos resultados en promedios:

```python
df_top20['Average'] = df_top20['Fatalities_Sum'] / df_top20['Onboard_Total_Sum']
df_top20[['Aircaft_Model', 'Onboard_Total_Sum', 'Fatalities_Sum', 'Average']].sort_values(by='Average', ascending=False)
```

Se puede ver claramente que si bien el **Douglas C-47-DL (DC-3)** es el modelo de aeronave que más casos registra y mayor cantidad de fallecidos suma, el **Boeing 727** es el que más personas a bordo suma y el que menor cantidad de fallecidos posee, ya sea en suma como en promedio. Seguiremos analizando éstos casos en particular por intermedio de 2 nuevas preguntas:

- ¿Cuáles son los tipos de accidentes o incidentes?
- ¿Cuáles son los tipos de vuelos?

Comenzamos por una visualización simple para averiguar cuál de ambas aeronaves tiene mayor capacidad:

```python
filtro_B727 = df[df['Aircaft_Model'] == 'Boeing 727']
filtro_B727 = filtro_B727.sort_values(by='Onboard_Total', ascending=False)
print(filtro_B727[['Aircaft_Model', 'Onboard_Total']])
```

```python
filtro_DC3 = df[df['Aircaft_Model'] == 'Douglas C-47-DL (DC-3)']
filtro_DC3 = filtro_DC3.sort_values(by='Onboard_Total', ascending=False)
print(filtro_DC3[['Aircaft_Model', 'Onboard_Total']])
```

Con las visualizaciones indicadas, en la cual se utilizan los valores de la columna 'Onboard_Total', filtrando el modelo de aeronave deseada y ordenando de mayor a menor, se puede ver que el modelo **Boeing 727** tiene en sus valores máximos de personas a bordo mayor capacidad que el modelo **Douglas C-47-DL (DC-3)**, a razón de casi el triple en algunos casos. Ésto marca la principal diferencia de por qué una aeronave que registra mayor cantidad de casos, suma menor cantidad de personas a bordo.

---

Procedemos a graficar las diferencias entre tipos de accidentes o incidentes y tipos de vuelos entre ambas aeronaves:

```python
data1 = filtro_B727['Incident_Category'].value_counts()
labels = data1.index

plt.figure(figsize=(10, 6))
plt.pie(data1, labels=labels, autopct='%1.1f%%', startangle=65)
plt.title('Tipo de Incidentes - Boeing 727')

data2 = filtro_DC3['Incident_Category'].value_counts()
labels = data2.index

plt.figure(figsize=(10, 6))
plt.pie(data2, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Tipo de Incidentes - Douglas (DC-3)')

plt.show()
```

En el primer gráfico se muestra una tendencia diferenciada entre ambos modelos:

- Boeing 727 posee un mayor porcentaje (97,5%) de casos de secuestros sin daños mayores en aeronaves. El resto está compuesto en su mayoría por otros hechos criminales, pero todos los casos resultan en aeronave con daños reparables.
- Douglas C-47-DL (DC-3) tiene un 84,1% de casos de accidentes, pero si bien diversifica un poco más el resto de casos, llegando a incluír hechos criminales, la mayoría de los mismos sufre además pérdida del casco de la aeronave. La gravedad de éste tipo de daños es la causa más común de fallecimientos.

```python
filtros_concat = pd.concat([filtro_B727, filtro_DC3])

plt.figure(figsize=(8, 6))
sns.countplot(x='Aircaft_Model', hue='Flight_Type', data=filtros_concat)
plt.xlabel('Modelo de Aeronave')
plt.ylabel('Cantidad de Registros')
plt.title('Cantidad de Registros por Modelo de Aeronave y Tipo de Vuelo')
plt.tight_layout()
plt.show()
```

En éste gráfico notamos que hay discrepancias en los datos, ya que sabemos por visualizaciones anteriores que el modelo **Douglas C-47-DL (DC-3)** presenta 459 registros, pero en el gráfico a priori no llega a los 150. Consideramos entonces los posibles valores nulos que pueda tener la columna 'Flight_Type':

```python
print(filtro_B727['Flight_Type'].unique())
print(filtro_DC3['Flight_Type'].unique())
```

Reescribimos el código para que incluya los nulos y se pueda contabilizar los registros de la forma más exacta posible:

```python
filtros_concat['Flight_Type'].fillna('Unknown', inplace=True)

plt.figure(figsize=(10, 6))
sns.countplot(x='Aircaft_Model', hue='Flight_Type', data=filtros_concat, palette='Set2')
plt.xlabel('Modelo de Aeronave')
plt.ylabel('Cantidad de Registros')
plt.title('Cantidad de Registros por Modelo de Aeronave y Tipo de Vuelo')

labels = [item.get_text() for item in plt.gca().get_yticklabels()]
labels[0] = 'NaN'
plt.gca().set_yticklabels(labels)

plt.tight_layout()
plt.show()
```

Notamos que la mayoría de los registros de el modelo **Douglas C-47-DL (DC-3)** figura con nulos en la columna 'Flight_Type', lo cual es un aspecto a mejorar sobre la base de datos cruda. No obstante, considerando los análisis previos de tipo de daño, tipo de accidente o incidente y sobretodo la capaciad de la aeronave, se podría deducir con alta probabilidad que la mayoría de éstos nulos corresponden a vuelos domésticos.

Para ello, utilizamos un último código que nos muestre la naturaleza o utilidad de la aeronave:

```python
count_nature = filtro_DC3['Aircaft_Nature'].value_counts()

# Crear un DataFrame con las columnas 'Aircaft_Model' y 'Aircaft_Nature' y la cantidad de registros
count_view = pd.DataFrame({
    'Aircaft_Nature': count_nature.index,
    'Cantidad de Registros': count_nature.values
})

visualizacion = count_view.sort_values(by='Cantidad de Registros', ascending=False)

print(visualizacion)
```

Notamos que tiene un número muy bajo en relación a vuelos domésticos y de pasajeros, lo que nos ayuda a reforzar la hipótesis planteada previamente de que la mayoría de los valores nulos de la columna 'Filght_Type' corresponden a vuelos domésticos.
