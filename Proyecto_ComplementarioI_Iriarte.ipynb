Modificamos los valores totales de la columna 'Onboard_Total' ya que tiene un formato repetitivo respecto a las 2 columnas anteriores y la posterior.
Nos será de mayor utilidad si sólo mostrara un valor numérico de total de pasajeros y tripulación para luego hacer cálculos precisos con respecto a la columna 'Fatalities', que también es tipo int64.

```python
def procesar_valor(valor):

    partes = valor.split('/ ')
    if len(partes) > 1:
        valor_final = partes[1]
    else:
        valor_final = partes[0]

    numeros = [int(s) for s in valor_final.split() if s.isdigit()]

    if len(numeros) > 0:
        return numeros[-1]
    else:
        return 0

df['Onboard_Total'] = df['Onboard_Total'].apply(procesar_valor)

df['Onboard_Total']
```

Las columnas 'Onboard_Crew' y 'Onboard_Passengers' también presentan formato object los cuales tienen datos específicos de personas a bordo y personas fallecidas en la misma instancia.

```python
df[['Onboard_Crew', 'Onboard_Passengers']]
```

Se aplica un código que convierta éstos datos, al igual que los anteriores, para utilizar las columnas en formato int. Para ello, se deja sólo el número de personas a bordo en cada columna extrayendo el número final de cada registro y en caso de no tener número, se asigna 0.

Además se crean 2 columnas nuevas llamadas 'Crew_Fatalities' y 'Pass_Fatalities' las cuales utilizan otro proceso de extracción de número, pero en éste caso el que se asigna luego de 'Fatalities: ' y y en caso de no tener número, se asigna 0.

```python
def extract_fatalities(text):
    if isinstance(text, str):
        match = re.search(r'Fatalities: (\d+)', text)
        if match:
            return int(match.group(1))
    return 0

df['Onboard_Crew'] = df['Onboard_Crew'].apply(lambda x: int(re.findall(r'\d+', str(x))[-1]) if re.findall(r'\d+', str(x)) else 0)
df['Onboard_Passengers'] = df['Onboard_Passengers'].apply(lambda x: int(re.findall(r'\d+', str(x))[-1]) if re.findall(r'\d+', str(x)) else 0)

df['Crew_Fatalities'] = df['Onboard_Crew'].apply(extract_fatalities)
df['Pass_Fatalities'] = df['Onboard_Passengers'].apply(extract_fatalities)
```

Vista final de las columnas modificadas y las nuevas creadas.

```python
df[['Onboard_Crew', 'Crew_Fatalities', 'Onboard_Passengers', 'Pass_Fatalities']]
```

Agregamos la columna 'Flight_Type' la cual define con una función condicional si el país de la columna 'Departure_Airport' coincide con el país de la columna 'Destination_Airport' entonces devolverá el valor 'Domestic', caso contrario el valor será 'International' y se asignará 'nan' a los registros que poseen datos como '?' y '-'.

```python
def extract_country(text):
    match = re.search(r',\s([^,]+)$', text)
    if match:
        return match.group(1)
    else:
        return np.nan
def determine_flight_type(row):
    departure_country = extract_country(row['Departure_Airport'])
    destination_country = extract_country(row['Destination_Airport'])
    if pd.isna(departure_country) or pd.isna(destination_country):
        return np.nan
    if departure_country == destination_country:
        return 'Domestic'
    else:
        return 'International'
df['Flight_Type'] = df.apply(determine_flight_type, axis=1)

df['Flight_Type']
```

Cobre la columna 'Aircaft_First_Flight' se identifican diversidad de formatos de valores, donde a veces posee solo 4 dígitos que indican un año, en otros casos poseen una fecha completa y también pueden tener una referencia de edad por meses y/o años de antigüedad entre parentesis.

```python
df['Aircaft_First_Flight']
```

Ingresamos un código de cálculo que nos permita extraer sólo los primeros 4 dígitos que indiquen el año del primer vuelo y luego restar dicho valor - el número existente en la columna 'Year' para devolver la edad de la aeronave al momento del suceso, en una columna nueva llamada 'Aircraft_Age'. Se agrega un condifinal que devuelva valor 0 en caso de que la columna 'Aircaft_First_Flight' no posea valores, sea NaN o tenga valor 0, para que en ésos casos entonces no realice la resta, sino que devuelva valor 0 en la columna nueva (de ésta manear evitamos posibles errores en los cuales, por ejemplo, veamos aeronaves con edad 1962).

```python
def calcular_edad(row):
    first_flight_year = str(row['Aircaft_First_Flight'])[:4]
    if first_flight_year.isdigit() and int(first_flight_year) != 0:
        return row['Year'] - int(first_flight_year)
    else:
        return 0

df['Aircraft_Age'] = df.apply(calcular_edad, axis=1).astype(int)

df['Aircraft_Age']
```

Corremos un código de los valores más altos obtenidos en la nueva columna, con el fin de detectar posibles errores en caso de anomalías.

Llamamos a las librerías con las que vamos a trabajar.

- Procedemos a reducir dimensionalidad con Bacward Elimination (se probó con Forward y Stepwise, pero simplemente no arroja resultados).
- Generamos los códigos de entrenamiento y testeo.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
from matplotlib import pyplot as plt
```

```python
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break 
    return features
```

```python
backward_elimination(X,y)
```
